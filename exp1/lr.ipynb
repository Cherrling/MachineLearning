```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
```

```python
samples = np.array([[1,0],[0,1],[2,2],[3,3],[5,5],[6,6]])

# 尝试更多的样本，随机性强一点
samples = np.array([[1,0],[0,1],[2,2],[3,3],[5,5],[6,6],[1,1],[2,1],[3,2],[6,2],[5,3],[6,4],[1,5],[2,6],[3,5],[5,6]])
```

```python


samples_x= samples[:,0]
samples_y= samples[:,1]

regr = linear_model.LinearRegression()
regr.fit(samples_x.reshape(-1,1), samples_y)

pre_x=np.linspace(np.min(samples_x)-1, np.max(samples_x)+1, 10, endpoint=True)
pre_y=regr.predict(pre_x.reshape(-1,1))
print(pre_y)

plt.scatter(samples_x, samples_y, s=50,color='black')
plt.plot(pre_x, pre_y, color='blue', linewidth=3)
plt.show()
```
